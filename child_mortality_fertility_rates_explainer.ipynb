# The probabilistic implications of the high pre-1800 global child mortality and fertility rates 

Before 1800, the global child mortality rate was [43%](https://ourworldindata.org/child-mortality) and global fertility rate (average number of children born per woman) was around [6](https://www.gapminder.org/topics/babies-per-woman/) [1,2]. One may wonder why women had so many children, far greater than the current global average of around 2.4 children born per woman. One possible reason is the lack of contraception, another is the potential usefulness of having many children to help out with subsistence farming, but there may be another reason. 

Parents may have had many children to ensure that at least one or two successfully survive past childhood. With a high child mortality rate, only having two children runs a large risk that one or both might die before turning five years old. As discussed by [Haines](https://www.ncbi.nlm.nih.gov/books/NBK233807/), there is some historical evidence for this theory [3]. However, exploring the probabilties behind ensuring at least two children survive into adulthood may lend even greater support.

```python
import child_mortality_fertility as cmf
```

```python
pre_1800_prob_6 = cmf.probabilities_num_alive(0.43, 6)
```

Here we can see the probability that a certain number of children will survive past 5 years old using the pre-1800 child mortality rate of 43% (before five years old) and fertility rate of 6 children per woman. You can see that it's most likely that about 3 or 4 children will survive past 5. It's highly unlikely that all six children will die (0.6%) and it's also highly unlikely that all six children will survive past five years old (3.4%). 

```python
cmf.plot_probabilities_num_alive(pre_1800_prob_6)
```

```python
pre_1800_1_to_6 = cmf.probabilities_num_alive_list(0.43, 6)
```

We can plot all six histograms showing the probabilities of each number of children surviving when having 1 through 6 six children.

```python
cmf.plot_probabilities_num_alive_list(pre_1800_1_to_6)
```

```python
pre_1800_ex_num_1_to_6 = cmf.expected_number_surviving_list(pre_1800_1_to_6)
```

Here, we plot the expected number of children surviving based on number of children born. This increases linearly in increments of 0.57. That is, a pre-1800 couple having n children could expect n * 0.57 to survive past five years old. For example, a pre-1800 couple having 5 children could most likely expect 3 to survive and 2 to die before the age of five years old.

```python
cmf.plot_expected_number_surviving_list(pre_1800_ex_num_1_to_6)
```

What may truly drive the large number of children born is guaranteeing that at least one or two survive past five years old. Here we plot the probability that at least 2 children survive past five years old based on the number of children born. That is, a couple having only 2 children faces a 32% chance that one or both will die before the age of five. That family can nearly double their probability of 'reproductive success' that is, at least two children successfully surviving past five years old, by having just one more child. The increases in the probability of reproductive success continue to increase but in diminshing amounts. At six children born, the couple has nearly guaranteed the probability of 'reproductive success', with a 94% of that at least two of their children will survive past five years old.

```python
prob_at_least_2_surv_1_to_6 = cmf.prob_at_least_n_survive_list(2, pre_1800_1_to_6)
cmf.plot_prob_at_least_n_survive_list(prob_at_least_2_surv_1_to_6)
```

Here we can see the diminishing marginal utility of having ever more children (in terms of obtaining the reproductive success of at least two children surviving past five). There are huge gains in the probability of reproductive success when going from 2 to 5 children, but past having 5 children, the gains become less and less, while the difficulty of raising the additional children likely remained relatively constant or possibly increased. At 6 children per woman, the value of increasing the probability of reproductive success from 94% to 97% may have been less than or equal to the cost of raising giving birth to or raising additional 7th child. Thus at that point, it may not have been rational to have the 7th child.
